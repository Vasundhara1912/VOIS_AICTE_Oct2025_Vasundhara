```python
# Convert potentially categorical columns back to object type before filling NaNs
df['neighbourhood group'] = df['neighbourhood group'].astype('object')
df['neighbourhood'] = df['neighbourhood'].astype('object')
df['room type'] = df['room type'].astype('object')

#replacing null values with 0 for numerical and datetime columns
for col in df.columns:
    if df[col].dtype != 'object':
        df[col].fillna(0, inplace=True)
```
